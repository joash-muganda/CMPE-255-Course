## **Convert Pdf to  txt**

```python
import PyPDF2
import os

def extract_text_from_pdf(pdf_path):
    text = ""
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            for page in reader.pages:
                if page.extract_text():
                    text += page.extract_text() + "\n"
    except Exception as e:
        text = f"Error reading {pdf_path}: {e}"
    return text

def batch_convert_pdf_to_text():
    for filename in os.listdir():
        if filename.endswith('.pdf'):
            text = extract_text_from_pdf(filename)

            # Output the text to a .txt file
            output_filename = filename.replace('.pdf', '.txt')
            with open(output_filename, 'w', encoding='utf-8') as text_file:
                text_file.write(text)
            print(f"Processed {filename}")

# Call the function
batch_convert_pdf_to_text()
```

# **Extract Embeddings for Each Document Using an LLM**

### **transformers**

```python
!pip install -U sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
import os

# Initialize the sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

def generate_embedding(text):
    return model.encode(text)

# Function to read text files and generate embeddings
def process_documents():
    embeddings = []
    document_names = []
    for filename in os.listdir():
        if filename.endswith('.txt'):
            with open(filename, 'r', encoding='utf-8') as file:
                text = file.read()
                embeddings.append(generate_embedding(text))
                document_names.append(filename)
    return document_names, embeddings

# Process the documents and get their embeddings
document_names, document_embeddings = process_documents()

# Outputs
print("Embeddings generated for the following documents:")
for name in document_names:
    print(name)

```

### **Apply a Clustering Algorithm**

```python
from sklearn.cluster import KMeans
import numpy as np

# Convert list of embeddings to a numpy array
embeddings_array = np.array(document_embeddings)

# Choose the number of clusters
# This is an arbitrary choice for demonstration; adjust based on your analysis
num_clusters = 5

# Apply K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(embeddings_array)

# Output the cluster assignments
for doc_name, cluster_id in zip(document_names, clusters):
    print(f"{doc_name}: Cluster {cluster_id}")
```

```python
from transformers import BertModel, BertTokenizer

model_name = 'bert-base-uncased'  # or 'roberta-base'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
from transformers import BertModel, BertTokenizer
import torch

# Initialize the model and tokenizer
model_name = 'bert-base-uncased'  # or 'roberta-base' or any other model you prefer
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
def generate_embedding(text):
    # Ensure the input is a string
    if not isinstance(text, str):
        raise ValueError("Input text must be a string.")

    inputs = tokenizer(text, return_tensors='pt', truncation=True, max_length=512)
    outputs = model(**inputs)
    cls_embedding = outputs.last_hidden_state[:, 0, :]  # [CLS] token embedding
    return cls_embedding.detach().numpy()
```

```python
document_texts = []
for filename in os.listdir():
    if filename.endswith('.txt'):
        with open(filename, 'r', encoding='utf-8') as file:
            document_texts.append(file.read())
```

```python
bert_embeddings = [generate_embedding(text) for text in document_texts if isinstance(text, str)]
```

```python
# Print the number of documents processed
print(f"Number of documents processed: {len(bert_embeddings)}")

# Optionally, print the shape of each embedding vector (they should all be the same)
for i, embedding in enumerate(bert_embeddings):
    print(f"Embedding shape for document {i}: {embedding.shape}")
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Convert the list of embeddings to a numpy array
embeddings_array = np.vstack([embedding.squeeze() for embedding in bert_embeddings])

# Choose the number of clusters
num_clusters = 5  # This is an arbitrary choice; adjust as needed

# Apply K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(embeddings_array)

# Output the cluster assignments
for i, cluster_id in enumerate(clusters):
    print(f"Document {i}: Cluster {cluster_id}")
```
